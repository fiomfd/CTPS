#### 2-1. Two parallel lines and a transversal line

Suppose that two parallel lines $\ell_1$ and $\ell_2$ on the plane. 
Pick up a point A on $\ell_1$ and a point B on $\ell_2$ arbitrary, 
and consider the line $L:=\overline{\text{AB}}$ connecting A and B. 

Note that vertically opposite angles are equal to each other. We observe the following:

- Each pair of corresponding angles on the same side of the intersecting transversal are equal to each other.

- Each pair of alternate angles around the transversal are equal to each other.

- The sum of co-interior angles on the same side of an intersecting transversal is $180$ degrees.


#### Corresponding angles
We can choose $\text{B}=(b,0)$ for $-4\leqq b \leqq 4$ with step size $0.5$.

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact

def slope_intercept(b=-2):
    if b<0:
       T=np.atan(-5/b)
    else:
       T=np.pi/2+np.atan(b/5) 
    
    t = np.linspace(0, T, 20)
    u1=np.cos(t)
    v1=np.sin(t)+5
    u2=np.cos(t)+b
    v2=np.sin(t)
    
    px1 = np.array([0])
    py1 = np.array([5])
    px2 = np.array([b])
    py2 = np.array([0])
    plt.title("Corresponding angles")
    plt.axhline(y=5, linewidth=1, color='cyan', linestyle='-', label='l1')
    plt.axhline(y=0, linewidth=1, color='magenta', linestyle='-',label='l2')
    plt.plot([-b,2*b], [10,-5], color='lightgreen', label='L')
    plt.scatter(px1, py1, color='red', label='A')
    plt.scatter(px2, py2, color='purple', label='B')
    plt.plot(u1, v1, linewidth=3, color='blue')
    plt.plot(u2, v2, linewidth=3, color='blue')
    plt.xlim([-6,6])
    plt.ylim([-2,7])
    plt.axis('off')
    plt.grid(False)
    plt.legend()
    plt.show()

interact(slope_intercept, b=(-4, 4, 0.5));
```

#### 2-2. Sum of interior angles of polygones

#### Triangles
We first observe that the prpperties of alternative angles prove that the sum of interior angles of a triangle is $180$ degrees. We can slide the top angle A in the horizontal direction for $-4\leqq a \leqq 4$ with step size $0.5$.

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact

def slope_intercept(a=0):
    if a<=-3:
        S=np.pi/2+np.atan(-(3+a)/5)
        T=np.atan(5/(2-a))
    elif a<2:
        S=np.atan(5/(a+3))
        T=np.atan(5/(2-a))
    else:
        S=np.atan(5/(a+3))
        T=np.pi/2-np.atan((2-a)/5)
   
    s = np.linspace(0, S, 20)
    u1=-np.cos(s)+a
    v1=-np.sin(s)+5
    u2=np.cos(s)-3
    v2=np.sin(s)

    t = np.linspace(0, T, 20)
    u3=np.cos(t)+a
    v3=-np.sin(t)+5
    u4=-np.cos(t)+2
    v4=np.sin(t)
    
    px1 = np.array([a])
    py1 = np.array([5])
    px2 = np.array([-3])
    py2 = np.array([0])
    px3 = np.array([2])
    py3 = np.array([0])
    plt.title("Sum of interior angles of a triangle")
    plt.axhline(y=5, linewidth=1, color='cyan', linestyle='-', label='l1')
    plt.axhline(y=0, linewidth=1, color='cyan', linestyle='-',label='l2')
    plt.plot([a,-3], [5,0], linewidth=3, color='lightgreen')
    plt.plot([a,2], [5,0], linewidth=3, color='lightgreen')
    plt.plot([-3,2], [0,0], linewidth=3, color='lightgreen')
    plt.plot(u1, v1, linewidth=3, color='blue')
    plt.plot(u2, v2, linewidth=3, color='blue')
    plt.plot(u3, v3, linewidth=3, color='red')
    plt.plot(u4, v4, linewidth=3, color='red')
    plt.scatter(px1, py1, color='red', label='A')
    plt.scatter(px2, py2, color='purple', label='B')
    plt.scatter(px3, py3, color='orange', label='C')
    plt.xlim([-6,6])
    plt.ylim([-2,6])
    plt.axis('off')
    plt.grid(False)
    plt.legend()
    plt.show()

interact(slope_intercept, a=(-4, 4, 0.5));
```

#### $N$-sided polygons

Suppose $N=3,4,5,\dotsc$. Any $N$-sided polygon is splitted into $m$ convex polygon(s) with some $m=1,\dotsc,N-2$. We now observe that any $N$-sided convex polygon is splitted into $N-2$ triangles for $N=3,4,5,\dotsc,10$. So the sum of interior agles of any general $N$-sided polygon is $180\times(N-2)$ degrees. 

We create an $N$-sided convex polygon by picking up $N$ points on a circle randomly.


```python
import numpy as np
import matplotlib.pyplot as plt
import random
from ipywidgets import interact

def polygon1(N=7):
    lst = list(range(1, 17))
    lst = random.sample(lst,N-1)
    lst.sort()
    lst0 = [0]
    lst = lst0 + lst + lst0
    angles0 = np.array(lst)
    angles=angles0*20

    xs = 3*np.sin(np.deg2rad(angles))
    ys = 3*np.cos(np.deg2rad(angles))
    
    plt.title('Triangulated N-polygon')
    plt.plot(xs,ys, linewidth=2, color='lightgreen')
    plt.axis('off')
    for i in range(2,N-1):
        plt.plot((xs[0],xs[i]),(ys[0],ys[i]), linewidth=0.5, color='blue', linestyle='--')   
    plt.show()

interact(polygon1, N=(3, 10, 1));
```

#### 2-3. Sum of exterior angles of convex polygons

We observe that the sum of exterior angles of any convex polygons is $360$ degrees.

We create an $N$-sided convex polygon by picking up $N$ points on a circle randomly, and shrink it ($r \rightarrow 0$). 


```python
import numpy as np
import matplotlib.pyplot as plt
import random
from ipywidgets import interact

def vertex2(N=7):
    lst = list(range(1, 17))
    lst = random.sample(lst,N-1)
    lst.sort()
    lst0 = [0]
    lst = lst0 + lst + lst0
    angles0 = np.array(lst)
    angles=angles0*20
    xs = 3*np.sin(np.deg2rad(angles))
    ys = 3*np.cos(np.deg2rad(angles))
    rho=np.zeros(N+1)
    u=np.zeros(N)
    v=np.zeros(N)
    for i in range(0,N):
        rho[i]=np.sqrt((xs[i+1]-xs[i])**2+(ys[i+1]-ys[i])**2)
        u[i]=(xs[i+1]-xs[i])/rho[i]
        v[i]=(ys[i+1]-ys[i])/rho[i]
    
    def polygon2(r=1):
        rxs=r*xs
        rys=r*ys
        
        plt.title('Sum of exterior angles of a convex N-polygon')
        plt.plot(rxs,rys, linewidth=2, color='lightgreen')
        plt.xlim([-4,4])
        plt.ylim([-4,4])
        plt.axis('off')
        for i in range(0,N):
            plt.plot((rxs[i],rxs[i+1]+u[i]),(rys[i],rys[i+1]+v[i]), linewidth=0.5, color='blue', linestyle='--')   
            
        plt.show()
    interact(polygon2, r=(0,1,0.1));

interact(vertex2, N=(3,10,1));
```

### 3. Basic isometries on the plane

We observe the rigid motion of a triangle on the plane:  

- Translation
- Rotation
- Reflection

These are isometries on the plane, and any isometry on the plane is a composition of finite number of these three kinds of operations. 

#### 3-1. Translation

We observe the translation of a given triagle. 

1. Choose the direction $0 \sim 360$ degrees.
2. Choose the length of translation $-10 \sim 10$.


```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact

def translation(theta=30,r=7):
        
    plt.title('Translation of a triangle')
    plt.plot([0,2],[0,2], linewidth=3, color='lightgreen')
    plt.plot([0,3],[0,-3], linewidth=3, color='cyan')
    plt.plot([2,3],[2,-3], linewidth=3, color='magenta')
    plt.xlim([-10,15])
    plt.ylim([-14,12])
    plt.axis('off')

    plt.plot([0+r*np.cos(np.deg2rad(theta)),2+r*np.cos(np.deg2rad(theta))],
             [0+r*np.sin(np.deg2rad(theta)),2+r*np.sin(np.deg2rad(theta))],
             linewidth=3, color='lightgreen')
    plt.plot([0+r*np.cos(np.deg2rad(theta)),3+r*np.cos(np.deg2rad(theta))],
             [0+r*np.sin(np.deg2rad(theta)),-3+r*np.sin(np.deg2rad(theta))],
             linewidth=3, color='cyan')
    plt.plot([2+r*np.cos(np.deg2rad(theta)),3+r*np.cos(np.deg2rad(theta))],
             [2+r*np.sin(np.deg2rad(theta)),-3+r*np.sin(np.deg2rad(theta))],
             linewidth=3, color='magenta')

    plt.plot([0,r*np.cos(np.deg2rad(theta))],
             [0,r*np.sin(np.deg2rad(theta))],
             linewidth=0.5, color='blue', linestyle='--')
    plt.plot([2,2+r*np.cos(np.deg2rad(theta))],
             [2,2+r*np.sin(np.deg2rad(theta))],
             linewidth=0.5, color='blue', linestyle='--')
    plt.plot([3,3+r*np.cos(np.deg2rad(theta))],
             [-3,-3+r*np.sin(np.deg2rad(theta))],
             linewidth=0.5, color='blue', linestyle='--')

    plt.show()

interact(translation, theta=(0,360,1), r=(-10,10,0.1));
```

#### 3-2. Rotation
We observe the translation of a given triagle. 

1. Choose the center of rotation $(a,b) \in [-3,3]\times[-3,3]$.
2. Choose the angle of rotation ($0 \sim 360$ degrees).

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact

def translation(a=-3,b=3,theta=210):
    COSA=-1-a
    SINA=0-b
    COSB=1-a
    SINB=2-b
    COSC=2-a
    SINC=-3-b

    t = np.linspace(0, 360, num = 361)
    moveax = a + COSA*np.cos(np.deg2rad(t)) - SINA*np.sin(np.deg2rad(t))
    moveay = b + SINA*np.cos(np.deg2rad(t)) + COSA*np.sin(np.deg2rad(t))
    movebx = a + COSB*np.cos(np.deg2rad(t)) - SINB*np.sin(np.deg2rad(t))
    moveby = b + SINB*np.cos(np.deg2rad(t)) + COSB*np.sin(np.deg2rad(t))
    movecx = a + COSC*np.cos(np.deg2rad(t)) - SINC*np.sin(np.deg2rad(t))
    movecy = b + SINC*np.cos(np.deg2rad(t)) + COSC*np.sin(np.deg2rad(t))
         
    plt.title('Rotation of a triangle')
    plt.aspectratio=1
    plt.plot([-1,1],[0,2], linewidth=3, color='lightgreen')
    plt.plot([-1,2],[0,-3], linewidth=3, color='cyan')
    plt.plot([1,2],[2,-3], linewidth=3, color='magenta')
    plt.xlim([-17,21])
    plt.ylim([-15,16])
    plt.axis('off')
    plt.scatter(np.array([a]), np.array([b]), color='red', label='(0,0)')

    plt.plot([a+COSA*np.cos(np.deg2rad(theta))-SINA*np.sin(np.deg2rad(theta)),
             a+COSB*np.cos(np.deg2rad(theta))-SINB*np.sin(np.deg2rad(theta))],
             [b+SINA*np.cos(np.deg2rad(theta))+COSA*np.sin(np.deg2rad(theta)),
             b+SINB*np.cos(np.deg2rad(theta))+COSB*np.sin(np.deg2rad(theta))],
             linewidth=3, color='lightgreen')
    plt.plot([a+COSA*np.cos(np.deg2rad(theta))-SINA*np.sin(np.deg2rad(theta)),
             a+COSC*np.cos(np.deg2rad(theta))-SINC*np.sin(np.deg2rad(theta))],
             [b+SINA*np.cos(np.deg2rad(theta))+COSA*np.sin(np.deg2rad(theta)),
             b+SINC*np.cos(np.deg2rad(theta))+COSC*np.sin(np.deg2rad(theta))],
             linewidth=3, color='cyan')
    plt.plot([a+COSB*np.cos(np.deg2rad(theta))-SINB*np.sin(np.deg2rad(theta)),
             a+COSC*np.cos(np.deg2rad(theta))-SINC*np.sin(np.deg2rad(theta))],
             [b+SINB*np.cos(np.deg2rad(theta))+COSB*np.sin(np.deg2rad(theta)),
             b+SINC*np.cos(np.deg2rad(theta))+COSC*np.sin(np.deg2rad(theta))],
             linewidth=3, color='magenta')
 
    plt.plot(moveax[0:theta], moveay[0:theta], linewidth=0.5, color='blue', linestyle='--')
    plt.plot(movebx[0:theta], moveby[0:theta], linewidth=0.5, color='blue', linestyle='--')
    plt.plot(movecx[0:theta], movecy[0:theta], linewidth=0.5, color='blue', linestyle='--')     
    
    plt.show()

interact(translation, a=(-5,5,0.1), b=(-5,5,0.1), theta=(0,360,1));
```

### 4. Affine linear functions $\ $ $y=ax+b$  

We observe basic properties of affine linear functions:

- Slopes and intercepts
- Parallel translations

#### 4-1. Slopes and intercepts
We observe the graph of an affine function $y=ax+b$ for various slope $a$ and intercept $b$. This is the line joining points $(0,b)$ and $(1,a+b)$.  

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact

def slope_intercept(a=1,b=1):
    x = np.linspace(-5, 5, num = 101)
    y = a*x+b
    px1 = np.array([0])
    py1 = np.array([b])
    px2 = np.array([1])
    py2 = np.array([a+b])
    plt.title("y=ax+b")
    plt.axhline(y=0, linewidth=1, color='black', linestyle='-')
    plt.axvline(x=0, linewidth=1, color='black', linestyle='-')
    plt.axvline(x=1, linewidth=0.5, color='black', linestyle='--')
    plt.plot(x, y, color='cyan')
    plt.scatter(px1, py1, color='red', label='(0,b)')
    plt.scatter(px2, py2, color='magenta', label='(1,a+b)')
    plt.ylim([-11,11])
    plt.xlabel("x")
    plt.ylabel("y")
    plt.xticks([0,1])
    plt.yticks([0])
    plt.grid(False)
    plt.legend()
    plt.show()

interact(slope_intercept, a=(-5, 5, 0.1), b=(-5, 5, 0.1));
```

#### 4-2. Parallel transport of $y=ax$
We observe the parallel transport of the graph of a linear function $y=ax$ under the translation $(0,0) \mapsto (p,q)$ for various $a,p,q$. 

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact

def parallel_transport(a=1,p=1,q=-1):
    x = np.linspace(-5, 5, num = 101)
    y = a*x
    ypq = a*(x-p)+q
    px1 = np.array([0])
    py1 = np.array([0])
    px2 = np.array([p])
    py2 = np.array([q])
    plt.title("Parallel transport of y=ax+b")
    plt.axhline(y=0, linewidth=1, color='black', linestyle='-')
    plt.axhline(y=q, linewidth=0.5, color='black', linestyle='--')
    plt.axvline(x=0, linewidth=1, color='black', linestyle='-')
    plt.axvline(x=p, linewidth=0.5, color='black', linestyle='--')
    plt.plot(x, y, color='cyan', label='y=ax')
    plt.plot(x, ypq, color='lightgreen', label='y=a(x-p)+q')
    plt.scatter(px1, py1, color='red', label='(0,0)')
    plt.scatter(px2, py2, color='magenta', label='(p,q)')
    plt.ylim([-11,11])
    plt.xlabel("x")
    plt.ylabel("y")
    plt.xticks([0,1])
    plt.yticks([0])
    plt.grid(False)
    plt.legend()
    plt.show()

interact(parallel_transport, a=(-5, 5, 0.1), p=(-5, 5, 0.1), q=(-5, 5, 0.1));
```
